```python
# Columns for maximum analysis

columns = [
    'brent_usd_barrel',
    'wti_usd_barrel',
    'dubai_usd_barrel',
    'us_gas_avg_gallon',
    'us_diesel_avg_gallon',
    'strait_hormuz_daily_ships'
]

# Clear names for output

column_names = {
    'brent_usd_barrel' : 'Brent Crude Oil Price',
    
    'wti_usd_barrel' : 'WTI Crude Oil Price',
    
    'dubai_usd_barrel' : 'Dubai Crude Oil Price',
    
    'us_gas_avg_gallon' : 'U.S. Average Gasoline Price',
    
    'us_diesel_avg_gallon' : 'U.S. Average Diesel Price',
    
    'strait_hormuz_daily_ships' : 'Strait of Hormuz Daily Ships'
}

print("\n========== MAXIMUM VALUES ANALYSIS ==========\n")

# Loop through columns

for col in columns:
    
    max_value = df_oil[col].max()
    
    # Dollar formatting
    if col != 'strait_hormuz_daily_ships':
        
        print(f"Maximum {column_names[col]} : ${max_value}")
    
    else:
        
        print(f"Maximum {column_names[col]} : {max_value}")

print("\n=============================================")
```

```python
# Maximum prewar gas price in each region?
max_prewar_gas_price_per_zone = df_gas.groupby('region')['gas_price_prewar_feb27'].max().sort_values(ascending=False).reset_index()
print(max_prewar_gas_price_per_zone)
```

```python
# Maximum postwar gas price in each region?
max_postwar_gas_price_per_zone = df_gas.groupby('region')['gas_price_mar19_2026'].max().sort_values(ascending=False).reset_index()
print(max_postwar_gas_price_per_zone)
```

```python
# Highest Percentage increace since war in each region?
Highest_pct_inc_since_war_per_reg = df_gas.groupby('region')['pct_increase_since_war'].max().sort_values(ascending=False).reset_index()
print(Highest_pct_inc_since_war_per_reg)
```

```python
# Maximum price increase since war in each region?
max_price_inc_since_war_per_reg = df_gas.groupby('region')['price_increase_since_war'].max().sort_values(ascending=False).reset_index()
print(max_price_inc_since_war_per_reg)
```

```python
# State wise evaluation ? 
# State-wise gas price analysis

analysis = [
    {
        'title': 'State with Maximum Postwar Gas Price',
        'column': 'gas_price_mar19_2026',
        'ascending': False
    },
    
    {
        'title': 'State with Minimum Postwar Gas Price',
        'column': 'gas_price_mar19_2026',
        'ascending': True
    },
    
    {
        'title': 'State with Maximum Prewar Gas Price',
        'column': 'gas_price_prewar_feb27',
        'ascending': False
    },
    
    {
        'title': 'State with Minimum Prewar Gas Price',
        'column': 'gas_price_prewar_feb27',
        'ascending': True
    }
]

print("\n========== STATE-WISE GAS PRICE ANALYSIS ==========\n")

# Loop through analysis

for item in analysis:
    
    result = (
        df_gas
        .sort_values(
            by=item['column'],
            ascending=item['ascending']
        )[['state', item['column']]]
        .head(1)
    )
    
    print(f"\n{item['title']}:\n")
    
    print(result)

print("\n===================================================")
```

```python
# avg of gas price by region.
avg_price_inc_since_war_per_reg = df_gas.groupby('region')['price_increase_since_war'].mean().sort_values(ascending=False).reset_index()
avg_pct_increase_since_war = df_gas.groupby('region')['pct_increase_since_war'].mean().sort_values(ascending=False).reset_index()
avg_gas_price_prewar = df_gas.groupby('region')['gas_price_prewar_feb27'].mean().sort_values(ascending=False).reset_index()
avg_gas_price_postwar = df_gas.groupby('region')['gas_price_mar19_2026'].mean().sort_values(ascending=False).reset_index()
print('Average price icreace since war =')
print(avg_price_inc_since_war_per_reg)
print('Average percentage increament in gas since war =')
print(avg_pct_increase_since_war)
print('Average gas price before the war =')
print(avg_gas_price_prewar)
print('Average gas price after war=')
print(avg_gas_price_postwar)
```

```python
# Total ships cross by strait of hormuz before and after the war.
Total_ships_cross_strait_of_hormus = df_oil['strait_hormuz_daily_ships'].sum()
print('Total ships cross by strait of hormuz before and after the war = ',Total_ships_cross_strait_of_hormus)
```

```python
# Ships before and after the war.
# War start date
war_date = '2026-02-27'

# Split dataset
before_war = df_oil[df_oil['date'] <= war_date]

after_war = df_oil[df_oil['date'] > war_date]

# Sum ships
before_sum = before_war['strait_hormuz_daily_ships'].sum()

after_sum = after_war['strait_hormuz_daily_ships'].sum()

print("Total Ships Before War =", before_sum)

print("Total Ships After War =", after_sum)
```

```python
# percentage difference before and after
percentage_difference = (
    (after_sum - before_sum) / before_sum
) * 100

print(f"Percentage Difference = {percentage_difference:.2f}%")
```

```python
# oil price before and after the war and their percentge according to category wise.

# Define dates
war_start = '2026-02-27'
war_end = '2026-03-20'

# Columns to analyze
columns = [
    'brent_usd_barrel',
    'wti_usd_barrel',
    'dubai_usd_barrel',
    'us_gas_avg_gallon',
    'us_diesel_avg_gallon'
]

# Loop through columns
for col in columns:

    start_price = df_oil.loc[
        df_oil['date'] == war_start,
        col
    ].values[0]

    end_price = df_oil.loc[
        df_oil['date'] == war_end,
        col
    ].values[0]

    percentage_change = (
        (end_price - start_price) / start_price
    ) * 100

    print(f"\nColumn: {col}")

    print(f"Start Price = {start_price}")

    print(f"End Price = {end_price}")

    print(
        f"Percentage Change = {percentage_change:.2f}%"
    )
```

```python
# oil pricing according to the war day.
columns = [
    'brent_usd_barrel',
    'wti_usd_barrel',
    'dubai_usd_barrel'
]

for col in columns:
    
    max_price_war_day = (
        df_oil.groupby('war_day')[col]
        .max()
        .sort_values(ascending=False)
        .head(1)
    )
    
    print(f"\nMaximum value for {col}:\n")
    
    print(max_price_war_day)
```

```python
max_oil_phase_analysis = (
    df_oil.groupby('phase')[
        ['brent_usd_barrel',
         'wti_usd_barrel',
         'dubai_usd_barrel']
    ]
    .max()
    .round(2)
)
print('Maximum oil phase analysis')
print(max_oil_phase_analysis)
```

```python
avg_oil_phase_analysis = (
    df_oil.groupby('phase')[
        ['brent_usd_barrel',
         'wti_usd_barrel',
         'dubai_usd_barrel']
    ]
    .mean()
    .round(2)
)
print('Average oil phase analysis')
print(avg_oil_phase_analysis)
```

```python
min_oil_phase_analysis = (
    df_oil.groupby('phase')[
        ['brent_usd_barrel',
         'wti_usd_barrel',
         'dubai_usd_barrel']
    ]
    .min()
    .round(2)
)
print('Minimum oil phase analysis')
print(min_oil_phase_analysis)
```
